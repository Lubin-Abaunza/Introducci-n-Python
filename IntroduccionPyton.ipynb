<h1 style="color:red;"> Ejemplo 1 </h1>
Cuales variables del conjunto de datos Municipios tienen valores perdidos o missings?

```python
# crea la funcion para ver si hay missings
def missings(x):
            a = sum(x.isna())
            b = 100*np.mean(x.isna()) # 
            c = pd.Series({"Missings": a,"%Missings": b})
            return(c)
```

```python
# Aplica una función a lo largo del eje 0 (es decir, por columnas)

Municipios.apply(missings,axis=0)
# Municipios: Este es el DataFrame al que estás aplicando la función.
# .apply(missings, axis=0): El método apply de pandas aplica una función a lo largo del eje especificado. En este caso, missings es la función que estás aplicando a cada columna del DataFrame Municipios.
```

```python
# Aplica una función a lo largo del eje 1 (es decir, por filas)
Municipios.apply(missings,axis=1)
```

<h1 style="color:red;"> Ejemplo 2 </h1>
Se hace "limpieza" en los nombres de los municipios, es decir,

- Se eliminan caracteres especiales.

- Se eliminan espacios al inicio y al final.

- Se reemplazan espacios múltiples por sencillos.

- Se escriben en letras minúsculas excepto en la inicial de cada palabra.

```python
# Se crea una nueva varible llamada Municipios2 a la base de datos Municipios.
# Esta nueva varible corresponde a cojer la columna Municipio y "limpiarla"
Municipios["Municipio2"] = (Municipios.Municipio
                            .str.lower()
                            .str.replace("[^a-záéíóúüñ ]","", regex=True)
                            .str.replace(" +"," ", regex=True)
                            .str.strip()
                            .str.title()
                           ) 
# Municipios["Municipio2"]: Crea una nueva columna en el DataFrame Municipios llamada Municipio2.
# Municipios.Municipio: Selecciona la columna Municipio del DataFrame.
# .str.lower(): Convierte todos los caracteres de la columna Municipio a minúsculas.
# .str.replace("[^a-záéíóúüñ ]","", regex=True): Elimina todos los caracteres que no sean letras (incluyendo caracteres acentuados) o espacios.
# .str.replace(" +"," ", regex=True): Reemplaza cualquier secuencia de uno o más espacios por un solo espacio.
# .str.strip(): Elimina los espacios en blanco al inicio y al final de cada valor de la columna.
# .str.title(): Convierte la primera letra de cada palabra en mayúscula.


Municipios[["Municipio","Municipio2"]].tail(10) # Muestra las columna Municipio y Municipio2
```

<h1 style="color:red;"> Ejemplo 2 </h1>
Cálculo de algunas estadísticas descriptivas (personalizadas) para las variables numéricas en el conjunto de datos Municipios.

```python
# Crea un sumary para estadisticas descriptivas

def summary(y,include='numeric'):
    def numer(x):
        x2 = x.dropna() # Quita los NA
        a = x.isna().sum() # Suna la cantidad de NA que hay
        b = x2.mean() # Media
        c = x2.std() # Desviacion estandar
        d = c/b # Coeficiente de variacaion
        e = x2.skew() 
        f = x2.kurt()
        g = x2.min() # Minino
        h = np.quantile(x2,0.25)
        i = x2.median()
        j = np.quantile(x2,0.75)
        k = x2.max() # Maxino
        m = x2.max() - x2.min()
        n = j - h
        l = pd.Series({"Missings":a,"Mean":b,"St.dev":c,"Coef.var":d,"Skewness":e,
                       "Kurtosis":f,"Min":g,"%25":h,"Median":i,"%75":j,"Max":k,"Rango":m, "RIQ":n})
        return(l)
    if include == 'numeric': 
        numerics = y.select_dtypes(include='number')
        if len(numerics.columns) > 0: return(numerics.apply(numer)) 
        else: print('Error. There are not numeric-type columns!!')
    if include == 'object':    
        objects = y.select_dtypes(include='object')
        if len(objects.columns) > 0: return(objects.describe(include="object"))
        else: print('Error. There are not object-type columns!!')
```

```python
# Crea una funcion para comparar dos DataFrames
def comparacion(a,b):
    if len(a.select_dtypes(exclude='float').columns) > 0:
        Nf = a.select_dtypes(exclude='float').equals(b.select_dtypes(exclude='float'))
    else:
        Nf = True
    if len(a.select_dtypes(include='float').columns) > 0:
        f = a.select_dtypes(include='float').round(5).equals(b.select_dtypes(include='float').round(5))
    else:
        f = True
    if Nf==True & f==True:
        return(True)
    else:
        return(False)
```

<h1 style="color:red;"> Ejemplo 6 </h1>

Conjunto de datos con todos los departamentos, ordenados de mayor a menor con respecto a su población, que incluye la siguiente información: número de municipios, superficie, población, densidad poblacional, índice de ruralidad promedio ponderado por población, y una variable igual a "Grande" para los departamentos con población mayor a 1.5 millones de habitantes, "Pequeño" para departamentos con población menor a 300 mil habitantes, y "Mediano" para el resto.

```python
def wtmean(x):
    w = Municipios.loc[x.index,"Poblacion"]
    return(sum(x*w)/sum(w))
# Crea una funcion para calcula la media ponderada de una variable x, utilizando los valores de la columna Poblacion del DataFrame Municipios como pesos.
# Cuando lo use: Tengo que cambiar la columna "Poblacion" par la que mesecite y espesifiar el DataFrame que este usando

def cases(x):
    out = "Mediano"
    if sum(x) < 300000:  out= "Pequeño"
    if sum(x) > 1500000: out= "Grande"
    return(out) 
# Crea una funcion para hacer las categorias e la funcion x
# Si la quiero usar solo tengo que modificar los nombres de las categorias y los interbalos

e6 = (Municipios
      .groupby(["Dep","Departamento"])
      .agg(nmunicipios=("Depmun","count"),
           totpob=("Poblacion","sum"),
           totsup=("Superficie","sum"),
           Irural=("Irural",wtmean),
           totpobC=("Poblacion",cases)).reset_index()
      .eval('''denspob = totpob/totsup''') # Crea una otra columna con los resultados de las anteriores
      .sort_values(by=["totpob"],ascending=False).reset_index(drop=True)
      .loc[:,["Dep","Departamento","nmunicipios","totpob","totsup","Irural","denspob","totpobC"]] # Indica que varibles quiero
     )

e6.info()
e6.head()
```

<h4 style="color:blue;">9. ¿Cuál es el estado (CODESTAD y NOMESTAD) con la mayor densidad poblacional en municipios costeros?, es decir, ¿cual es el estado para el que la densidad poblacional en sus municipios costeros es la mayor del país?</h4>

```python
(MuniBR3
 .query("costero == 1")
 .groupby(["CODESTAD","NOMESTAD"])
 .agg(POB=("POB","sum"),
     SUP=("SUP","sum")).reset_index()
 .eval(""" DP = POB/SUP """)
 .sort_values(by=["DP"],ascending=False).reset_index(drop=True)
)
```

<h4 style="color:blue;">10 ¿Cuál es el estado (CODESTAD y NOMESTAD) con la mayor proporción de superficie en municipios costeros?, es decir, ¿cual es el estado para el que la proporción de su superficie que corresponde a municipios costeros es la mayor del país?</h4>

```python
SUPT = (MuniBR3
        .groupby(["CODESTAD","NOMESTAD"])
        .agg(SUPT=("SUP","sum")).reset_index()
        .sort_values(by=["SUPT"],ascending=False).reset_index(drop=True)
       )

SUPC = (MuniBR3
        .query("costero == 1")
        .groupby(["CODESTAD"])
        .agg(SUPC=("SUP","sum")).reset_index()
        .sort_values(by=["SUPC"],ascending=False).reset_index(drop=True)
       )

PRC  = (SUPT
        .merge(SUPC,left_on="CODESTAD",right_on="CODESTAD",how="left")
        .eval('''SUPC = SUPC.fillna(0).values''')
        .eval('''PRC = 100*(SUPC/SUPT)''')
        .drop(columns=["SUPC","SUPT"])
        .sort_values(by=["PRC"],ascending=False).reset_index(drop=True)
       )

PRC.head(10)

       
```

<h4 style="color:blue;">12. De los estados que no tienen ningún municipio costero, ¿cuál (CODESTAD y NOMESTAD) es el más homogéneo con respecto al tamaño, en número de habitantes, de los municipios que lo componen? Use el coeficiente de variación como medida de heterogeneidad.</h4>

```python
(MuniBR3
.groupby(["CODESTAD","NOMESTAD"])
.agg(mean=("POB","mean"),
     std=("POB","std"),
     costeros=("costero","sum")).reset_index()
.eval('''cv=std/mean''')     
.query("costeros==0") 
.sort_values(by=["cv"],ascending=[True]).reset_index(drop=True)
.head(2)
)
```
